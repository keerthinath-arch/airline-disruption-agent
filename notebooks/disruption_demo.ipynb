# Airline Disruption Forecasting Agent Demo

This notebook demonstrates a simplified version of the airline disruption forecasting system using signal weighting and decision logic.

```python
# Sample disruption scenario
disruption_input = {
    "weather": "moderate storm at ATL",
    "crew": "limited availability",
    "aircraft": "minor delays",
    "network": "incoming delays from JFK"
}

disruption_input
```

```python
# Risk scoring logic
def evaluate_risk(signals):
    score = 0
    if "storm" in signals["weather"]:
        score += 0.3
    if "limited" in signals["crew"]:
        score += 0.4
    if "delays" in signals["aircraft"]:
        score += 0.1
    if "delays" in signals["network"]:
        score += 0.2
    return round(score, 2)

risk_score = evaluate_risk(disruption_input)
risk_score
```

```python
# Determine dominant driver
def dominant_driver(signals):
    if "limited" in signals["crew"]:
        return "crew constraints"
    elif "storm" in signals["weather"]:
        return "weather conditions"
    elif "delays" in signals["network"]:
        return "network propagation"
    return "aircraft constraints"

driver = dominant_driver(disruption_input)
driver
```

```python
# Recommendation logic
def recommend_action(driver):
    if driver == "crew constraints":
        return "Reassign reserve crew to high-priority flights"
    elif driver == "weather conditions":
        return "Delay departures and adjust routing"
    elif driver == "network propagation":
        return "Protect key inbound connections"
    return "Swap aircraft if available"

recommendation = recommend_action(driver)
recommendation
```

```python
# Final output
output = {
    "risk_score": risk_score,
    "dominant_driver": driver,
    "recommendation": recommendation,
    "confidence": 0.72,
    "explanation": f"Primary disruption driver is {driver} based on input signals."
}

output
```
